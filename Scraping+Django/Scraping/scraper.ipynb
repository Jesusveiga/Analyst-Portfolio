```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import TimeoutException
import time
```

```python
# Configurar el webdriver para usar Chrome
webdriver_service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=webdriver_service)

# Abrir la URL en el navegador con el webdriver
url = "https://casa.sapo.pt/es-es/comprar-pisos-apartamentos/lisboa/"
driver.get(url)


# Crear una lista para almacenar la información de cada propiedad
properties_info = []

page_number = 1

try:
    while True:
        # Esperar a que la página se cargue completamente
        time.sleep(5)

        # Extraer la información deseada de los elementos HTML
        properties = driver.find_elements(By.CLASS_NAME, 'property-info')

        for property in properties:
            # Extraer el título
            title = property.find_element(By.CLASS_NAME, 'property-type').text.strip()

            # Extraer los metros cuadrados
            features = property.find_element(By.CLASS_NAME, 'property-features-text')
            square_meters = None
            under_construction = None
            if features:
                feature_text = features.text.strip()
                square_meters = feature_text.split('·')[1].strip() if '·' in feature_text else None
                under_construction = 'En construcción' in feature_text

            # Extraer el precio
            price = property.find_element(By.CLASS_NAME, 'property-price-value').text.strip()

            # Verificar si tiene garaje
            garage = False
            try:
                garage_element = property.find_element(By.XPATH, ".//div[@class='property-features-tag']/span[text()='Com Garagem']")
                garage = garage_element is not None
            except NoSuchElementException:
                pass

            # Extraer la ubicación
            location = property.find_element(By.CLASS_NAME, 'property-location').text.strip()

            # Añadir la información a la lista
            properties_info.append({
                'title': title,
                'square_meters': square_meters,
                'price': price,
                'under_construction': under_construction,
                'garage': garage,
                'location': location
            })

        page_number += 1

        # Buscar el enlace a la página siguiente y hacer clic en él
        next_page = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f"//a[@href='/es-es/comprar-pisos-apartamentos/lisboa/?pn={page_number}']")))
        driver.execute_script("arguments[0].scrollIntoView();", next_page)
        next_page.click()
except NoSuchElementException:
    # Si no hay enlace a la página siguiente, salir del bucle
    pass
except ElementClickInterceptedException:
    # Si el elemento no es clickeable, salir del bucle
    pass
finally:
    # Cerrar el navegador
    driver.quit()

    # Crear un DataFrame a partir de la lista de propiedades
    df_1 = pd.DataFrame(properties_info)

    # Imprimir el DataFrame
    print(df_1)
```
